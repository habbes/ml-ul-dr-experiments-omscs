## MADELON

```python
X_train = pd.read_hdf('data/madelon/madelon.hdf', 'X_train')
y_train = pd.read_hdf('data/madelon/madelon.hdf', 'y_train')
X_scaled = pp.scale(X_train)
```

```python
ica = dec.FastICA()
transformed = ica.fit(X_train).transform(X_train)
```

```python
plot_data = pd.DataFrame(transformed)
plot_data['class'] = y_train.values.ravel().astype(int)
sns.scatterplot(x=0, y=1, hue='class', data=plot_data)
```

## TELCO

```python
telX_train = pd.read_hdf('data/telco/telco.hdf', 'X_train')
tely_train = pd.read_hdf('data/telco/telco.hdf', 'y_train')
telX_scaled = pp.scale(telX_train)
```

```python
ica = dec.FastICA(max_iter=2000)
transformed = ica.fit(telX_scaled).transform(telX_scaled)
plot_data = pd.DataFrame(transformed)
plot_data['class'] = tely_train.values.ravel().astype(int)
sns.scatterplot(x=0, y=1, hue='class', data=plot_data)
```
